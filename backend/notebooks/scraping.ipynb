```python
from selenium import webdriver
pinterest_search_url = "https://www.pinterest.ca/search/pins/?q=snake%20fence&rs=sitelinks_searchbox"
```

```python
driver = webdriver.Chrome()
driver.get(pinterest_search_url)
```

```python
from selenium.webdriver.support import ui
wait = ui.WebDriverWait(driver, 10)
wait.until(driver.find_element_by_tag_name("body") is not None)
print("Loaded!")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, "html.parser")
soup
```

```python
images = soup.findAll("img", {"class": "GrowthUnauthPinImage__Image"})
images[0]
```

```python
class_name = "snake fence"
img_urls = []
for image in images:
    img_urls.append((image.get("src"), image.get("alt")))
img_urls
```

```python
len(img_urls)
```

Pinterest apparently limits API to 1000 calls per hour, hopefully the same for automated browser calls?

```python
import json
sample_urls = {
    "class": class_name,
    "urls": [u[0] for u in img_urls]
}
with open('sample_urls.json', 'w') as f:
    json.dump(sample_urls, f)
```
